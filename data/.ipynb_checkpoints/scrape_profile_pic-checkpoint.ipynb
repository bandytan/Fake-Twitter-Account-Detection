```python
df = pd.read_csv('all_users_2015.csv')
df['index'] = range(len(df))
df = pd.concat([df.iloc[:,-1],df.iloc[:,:-1] ], axis=1) # Move index to 1st column
df.head()
```

```python
df.shape
```

```python
def edit_url(url):
    starting = url[:16]
    pattern = 'a\d'
    if re.search(pattern, starting):
        replaced_pbs = re.sub(pattern, "pbs", starting)
    else:
        replaced_pbs = starting
    edited = replaced_pbs + url[16:]
    return edited
```

```python
# preprocess url to replace with working url

# Validate URL
url_pattern = "^https?:\\/\\/(?:www\\.)?[-a-zA-Z0-9@:%._\\+~#=]{1,256}\\.[a-zA-Z0-9()]{1,6}\\b(?:[-a-zA-Z0-9()@:%_\\+.~#?&\\/=]*)$"
re.match(url_pattern, 'https://uibakery.io') # Returns Match object
re.match(url_pattern, 'https:/uibakery.io') # Returns None

for index, row in df.iterrows():
    image_url = row['profile_image_url']
    df.loc[index, 'edited_url'] = edit_url(str(image_url))
#     result = re.match(url_pattern, image_url)
#     print(image_url[:16])
#     if not result: #If null or does not start with http or https
#         df['edited_url'] = ""
#     elif result:
#         df['edited_url'] = edit_url(result)
```

```python
failed_id = []
successful_id = []

for index, row in df.iterrows():
    id = row['id']
    imgURL = row['edited_url']
    try:
        conn = urllib.request.urlopen(imgURL)
#         except urllib.error.HTTPError as e:
#             # Return code error(e.g.404, 501, ...)
#             print('ID: {}, HTTPError: {}'.format(id, e.code))
#             error_404.append(id)
    except Exception as e:
        # Not an HTTP - specific error(e.g.connection refused)
#         print('ID: {}, Error'.format(id))
        failed_id.append(id)
    else :
        try:
            urllib.request.urlretrieve(imgURL, "batch2/{}.jpg".format(id))
        except Exception as e:
            failed_id.append(id)
        else:
            successful_id.append(id)
    print(index/5301)
```

```python
print(len(failed_id))
print(len(successful_id))
```

```python
with open("failed_id_batch2.json", "w") as output:
    json.dump(failed_id, output, indent=2) 
    
with open("successful_id_batch2.json", "w") as output:
    json.dump(successful_id, output, indent=2) 
```
