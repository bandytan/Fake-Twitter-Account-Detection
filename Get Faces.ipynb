```python
from facenet_pytorch import MTCNN, InceptionResnetV1
import torch
from torch.utils.data import DataLoader
from torchvision import datasets
import numpy as np
import pandas as pd
import os
from PIL import Image
import re
import pickle

device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
faces_base = "/Users/ivankoh/Downloads/new batch profile pics/"
```

```python
# get all image pathnames from base

# list to store files
res = []
res2 = []

# Iterate directory
for path in os.listdir(faces_base):
    # check if current path is a file
    if os.path.isfile(os.path.join(faces_base, path)):
        res.append(faces_base + path)
        res2.append(path)
```

```python
mtcnn = MTCNN(
    image_size=160, margin=0, min_face_size=20,
    thresholds=[0.6, 0.7, 0.7], factor=0.709, post_process=True,
    device=device
)

# Create an inception resnet (in eval mode):
resnet = InceptionResnetV1(pretrained='vggface2').eval().to(device)
```

```python
aligned = []
names = []
d={}
cannot_count = 0
can_count = 0
for i in range(len(res)):
    pic = res[i]
    id_name = re.match(r"[^\/\\]+(?=\.png|\.jpg)", res2[i]).group(0)
    try:
        img = Image.open(pic)
        x_aligned, prob = mtcnn(img, return_prob=True)
        print("Done", pic)
    except:
        print("Cannot:", pic)
    if x_aligned is not None:
        print('Face detected with probability: {:8f}'.format(prob))
        
        d[id_name] = 1
        aligned.append(x_aligned)
        #names.append(dataset.idx_to_class[y])
    else:
        print("x not aligned (no face detected):", prob)
        d[id_name] = 0
    
```

```python
print("have face:", len(list(filter(lambda x: x == 1,list(d.values())))))
print("no face:", len(list(filter(lambda x: x == 0,list(d.values())))))
```
